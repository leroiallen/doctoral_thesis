```python
%matplotlib inline
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi

panss = pd.read_excel('split_database_901_only_panss.xlsx')

#float64
panss = panss.astype('float64')

panss_p = panss[['panss_p1', 'panss_p2', 'panss_p3', 'panss_p4', 'panss_p5', 'panss_p6', 'panss_p7']]
panss_n = panss[['panss_n1', 'panss_n2', 'panss_n3', 'panss_n4', 'panss_n5', 'panss_n6', 'panss_n7']]
panss_g = panss[['panss_g1', 'panss_g2', 'panss_g3', 'panss_g4', 'panss_g5', 'panss_g6',
 'panss_g7', 'panss_g8', 'panss_g9', 'panss_g10', 'panss_g11',
 'panss_g12', 'panss_g13', 'panss_g14', 'panss_g15', 'panss_g16']]
panss_p_n_g = panss[['panss_p1', 'panss_p2', 'panss_p3', 'panss_p4', 'panss_p5', 'panss_p6', 'panss_p7', 
                      'panss_n1', 'panss_n2', 'panss_n3', 'panss_n4', 'panss_n5', 'panss_n6', 'panss_n7',
                     'panss_g1', 'panss_g2', 'panss_g3', 'panss_g4', 'panss_g5', 'panss_g6',
 'panss_g7', 'panss_g8', 'panss_g9', 'panss_g10', 'panss_g11',
 'panss_g12', 'panss_g13', 'panss_g14', 'panss_g15', 'panss_g16']]
panss_p_sum = panss['panss_p_sum']
panss_n_sum = panss['panss_n_sum']
panss_g_sum = panss['panss_g_sum']
panss_png_sum = panss[['panss_p_sum','panss_n_sum','panss_g_sum']]
panss_sum = panss['panss_sum']

panss_p_mean = panss_p.mean().values.tolist()
panss_p_mean += panss_p_mean[:1]
panss_p_std = panss_p.std().values.tolist()
panss_p_std += panss_p_std[:1]

panss_n_mean = panss_n.mean().values.tolist()
panss_n_mean += panss_n_mean[:1]
panss_n_std = panss_n.std().values.tolist()
panss_n_std += panss_n_std[:1]

N=7

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

plt.figure(figsize=(18,18))

ax1=plt.subplot(131, polar=True)
ax1.set_rlabel_position(0)
ax1.plot(angles, panss_p_mean, linewidth=3, color='skyblue')
ax1.fill_between(angles, np.array(panss_p_mean)-np.array(panss_p_std), np.array(panss_p_mean)+np.array(panss_p_std), color='skyblue', alpha=0.2)
plt.xticks(angles[:-1], ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7'], color='black', size=15)
plt.yticks([1,2,3,4,5,6], ['1','2','3','4','5','6'], color="grey", size=13)
plt.ylim(0,7)
plt.title('Positive Scale', size=17, position = (0.5,1.15))

ax2=plt.subplot(132, polar=True)
ax2.set_rlabel_position(0)
ax2.plot(angles, panss_n_mean, linewidth=3, color='salmon')
ax2.fill_between(angles, np.array(panss_n_mean)-np.array(panss_n_std), np.array(panss_n_mean)+np.array(panss_n_std), color='salmon', alpha=0.2)
plt.xticks(angles[:-1], ['N1', 'N2', 'N3', 'N4', 'N5', 'N6', 'N7'], color='black', size=15)
plt.yticks([1,2,3,4,5,6], ['1','2','3','4','5','6'], color="grey", size=13)
plt.title('Negative Scale', size=17, position = (0.5,1.15))
plt.ylim(0,7)

N=16

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

panss_g_mean = panss_g.mean().values.tolist()
panss_g_mean += panss_g_mean[:1]
panss_g_std = panss_g.std().values.tolist()
panss_g_std += panss_g_std[:1]

ax3=plt.subplot(133, polar=True)
ax3.set_rlabel_position(0)
ax3.plot(angles, panss_g_mean, linewidth=3, color='green')
ax3.fill_between(angles, np.array(panss_g_mean)-np.array(panss_g_std), np.array(panss_g_mean)+np.array(panss_g_std), color='green', alpha=0.2)
plt.xticks(angles[:-1], ['G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8', 'G9', 'G10', 'G11', 'G12', 'G13', 'G14', 'G15', 'G16'], color='black', size=15)
plt.yticks([1,2,3,4,5,6], ['1','2','3','4','5','6'], color="grey", size=13)
plt.ylim(0,7)
plt.title('General Psychopathology Scale', size=17, position = (0.5,1.15))

plt.savefig('panss_radar_chart.png',dpi=300)
```
