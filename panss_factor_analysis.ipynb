```python
%matplotlib inline
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

a=pd.DataFrame({'factor':range(1,31),
               'eigenvalue': [12.199, 5.654, 1.860, 1.179, 1.084, 0.788,0.657,0.653,0.585,
                              0.535,0.466,0.428,0.401,0.374,0.354,0.329,0.289,0.268,0.247,
                              0.212,0.187,0.185,0.174,0.169,0.158,0.133,0.130,0.121,0.110,0.073],
              'variance': [40.662,18.848,6.200,3.929,3.613,2.626,2.191,2.177,1.949,1.784,
                           1.555,1.427,1.335,1.247,1.179,1.096,0.962,0.893,0.823,0.705,0.624,
                           0.615,0.580,0.562,0.528,0.445,0.433,0.402,0.367,0.243],
               'accumulated_variance': [40.662,59.510,65.709,69.638,73.251,75.877,78.069,
                                        80.246,82.194,83.978,85.533,86.960,88.295,89.542,90.721,
                                        91.816,92.779,93.672,94.496,95.201,95.825,96.440,97.020,
                                        97.582,98.110,98.555,98.987,99.390,99.757,100.000]})


plt.plot(range(1,31),a['eigenvalue'],'o-',color='#a2cffe')
plt.xlabel('Factor')
plt.ylabel('Eigenvalue')
plt.xlim([0,31])


import warnings
warnings.filterwarnings("ignore")

df = pd.read_excel('panss_three_factor.xlsx')

plt.figure(figsize=[16,8])

for i in range(0,30):
    if df.ix[i,'factor_1'] > df.ix[i,'factor_2']:
        if df.ix[i,'factor_1'] > df.ix[i,'factor_3']:
            p1 = plt.bar(i*2, df.ix[i,'factor_1'],color='lightblue', width=1)
            if df.ix[i,'factor_2'] > df.ix[i,'factor_3']:
                p2 = plt.bar(i*2, df.ix[i,'factor_2'],color='salmon', width=1)
                p3 = plt.bar(i*2, df.ix[i,'factor_3'],color='pink', width=1)
            else:
                plt.bar(i*2, df.ix[i,'factor_3'],color='pink', width=1)
                plt.bar(i*2, df.ix[i,'factor_2'],color='salmon', width=1)
        else:
            plt.bar(i*2, df.ix[i,'factor_3'],color='pink', width=1)
            plt.bar(i*2, df.ix[i,'factor_1'],color='lightblue', width=1)
            plt.bar(i*2, df.ix[i,'factor_2'],color='salmon', width=1)
    else:
        if df.ix[i,'factor_1'] > df.ix[i,'factor_3']:
            plt.bar(i*2, df.ix[i,'factor_2'],color='salmon', width=1)
            plt.bar(i*2, df.ix[i,'factor_1'],color='lightblue', width=1)
            plt.bar(i*2, df.ix[i,'factor_3'],color='pink', width=1)
        else:
            if df.ix[i,'factor_2'] > df.ix[i,'factor_3']:
                plt.bar(i*2, df.ix[i,'factor_2'],color='salmon', width=1)
                plt.bar(i*2, df.ix[i,'factor_3'],color='pink', width=1)
                plt.bar(i*2, df.ix[i,'factor_1'],color='lightblue', width=1)
            else:
                plt.bar(i*2, df.ix[i,'factor_3'],color='pink', width=1)
                plt.bar(i*2, df.ix[i,'factor_2'],color='salmon', width=1)
                plt.bar(i*2, df.ix[i,'factor_1'],color='lightblue', width=1)

plt.legend((p1[0],p2[0],p3[0]), ('Factor 1', 'Factor 2', 'Factor 3'), loc='lower left',fontsize='x-large')
plt.yticks(fontsize='x-large')
plt.xticks(range(0,60,2), ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'N1', 'N2', 'N3', 'N4', 'N5', 'N6', 'N7', 
                           'G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8', 'G9', 'G10', 'G11', 'G12', 'G13', 'G14', 'G15', 'G16'], fontsize='x-large', rotation=30)
plt.plot((-2,60),(0.4,0.4), '--',color='grey', alpha=.5, linewidth=1)
plt.plot((-2,60),(0,0), '-',color='black', alpha=1, linewidth=1)
plt.xlim((-1.5,59.5))
plt.ylabel('Load',fontsize='xx-large')


plt.savefig('panss_factor_analysis_eigenvalue.png',dpi=300)
```
